```python
%pip install pandas

import pandas
```

```python
data = pandas.read_csv('data.csv')
data.head()
```

```python
data.dropna(inplace=True)
data.drop(columns=['source_id', 'source_name', 'author', 'url', 'url_to_image'], inplace=True)
data.head()
```

```python
#Отчистим от новостей не на английском языке
%pip install langdetect
from langdetect import detect

def detect_lang(text):
    try:
        return detect(text)
    except:
        return 'unknown'
    
data['lang'] = data['content'].apply(detect_lang)

data = data[data['lang'] == 'en']
data.drop(columns=['lang'], inplace=True)
data.head()
```

```python
data.describe(include='all')
```

```python
#Удалим все повторяющиеся новости
data.drop_duplicates(subset='title', inplace=True)
data.drop_duplicates(subset='description', inplace=True)
data.drop_duplicates(subset='content', inplace=True)

data.describe(include='all')
```
