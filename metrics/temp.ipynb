```python
import pandas as pd
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    roc_auc_score,
    precision_score,
    recall_score,
    f1_score
)
```

```python
prefixes = {
    'PlainLLM': [0,1,2],
    'NaiveRAG_gd': [0,1],
    'NaiveRAG_lc': [2],
    'HybridRAG_gd': [0],
    'HybridRAG_lc': [0,1,2],
    #'HybridRAG_CoT_lc': [0],
    'Community': [0],
}
```

```python
for prefix, runs in prefixes.items():
    cols = [f'{prefix}_{i}' for i in runs]
    df[f'{prefix}_mean'] = df[cols].mean(axis=1)
    df[f'{prefix}_std']  = df[cols].std(axis=1)

df
```

```python
print(f"Mean std for NaiveRAG_gd: {df['NaiveRAG_gd_std'].mean()}")
print(f"Mean std for NaiveRAG_lc: {df['NaiveRAG_lc_std'].mean()}")
print(f"Mean std for PlainLLM: {df['PlainLLM_std'].mean()}")
print(f"Mean std for HybridRAG_lc: {df['HybridRAG_lc_std'].mean()}")
print()
print(f"Mean mean for NaiveRAG_gd: {df['NaiveRAG_gd_mean'].mean()}")
print(f"Mean mean for NaiveRAG_lc: {df['NaiveRAG_lc_mean'].mean()}")
print(f"Mean mean for PlainLLM: {df['PlainLLM_mean'].mean()}")
print(f"Mean mean for HybridRAG_lc: {df['HybridRAG_lc_mean'].mean()}")
print(f"Mean mean for HybridRAG_gd: {df['HybridRAG_gd_mean'].mean()}")
#print(f"Mean mean for HybridRAG_CoT_lc: {df['HybridRAG_CoT_lc_mean'].mean()}")



```

```python
reg_metrics = {}
for prefix in prefixes:
    y_true = df['community_prediction']
    y_pred = df[f'{prefix}_mean']
    mse  = mean_squared_error(y_true, y_pred)
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = mse ** 0.5
    reg_metrics[prefix] = {'MSE': mse, 'RMSE': rmse, 'MAE': mae}

```

```python
clf_metrics = {}
for prefix in prefixes:
    y_true = df['answer'].astype(int)
    y_score = df[f'{prefix}_mean']
    auc = roc_auc_score(y_true, y_score)
    y_pred_label = (y_score >= 0.5).astype(int)
    prec = precision_score(y_true, y_pred_label, zero_division=0)
    rec  = recall_score(y_true, y_pred_label, zero_division=0)
    f1   = f1_score(y_true, y_pred_label, zero_division=0)
    clf_metrics[prefix] = {
        'ROC_AUC': auc,
        'Precision': prec,
        'Recall': rec,
        'F1': f1
    }
```

```python
print("Регрессионные метрики против community_prediction:")
for prefix, mets in reg_metrics.items():
    print(f"  {prefix}: MSE={mets['MSE']:.4f}, RMSE={mets['RMSE']:.4f}, MAE={mets['MAE']:.4f}")

print("\nКлассификационные метрики против answer:")
for prefix, mets in clf_metrics.items():
    print(f"  {prefix}: ROC_AUC={mets['ROC_AUC']:.4f}, "
          f"Precision={mets['Precision']:.4f}, Recall={mets['Recall']:.4f}, F1={mets['F1']:.4f}")







```

```python
# %% [markdown]
# **Шаг 7.** Рассчитаем взвешенные метрики регрессии и классификации:
# - Регрессия: MSE, RMSE, MAE с учётом sample_weight  
# - Классификация: ROC-AUC, Precision, Recall, F1 с учётом sample_weight  
# для двух видов весов: `total_predictors` и `mean_ACE`.

# %%
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    roc_auc_score,
    precision_score,
    recall_score,
    f1_score
)

weight_vars = ['total_predictors', 'mean_ACE']
weighted_reg_metrics = {w: {} for w in weight_vars}
weighted_clf_metrics = {w: {} for w in weight_vars}

for w in weight_vars:
    weights = df[w]
    for prefix in prefixes:
        # --- Регрессия ---
        y_true = df['community_prediction']
        y_pred = df[f'{prefix}_mean']
        mse_w = mean_squared_error(y_true, y_pred, sample_weight=weights)
        rmse_w = mse_w ** 0.5
        mae_w = mean_absolute_error(y_true, y_pred, sample_weight=weights)
        weighted_reg_metrics[w][prefix] = {
            'MSE': mse_w,
            'RMSE': rmse_w,
            'MAE': mae_w
        }
        # --- Классификация ---
        y_true_clf = df['answer'].astype(int)
        y_score = df[f'{prefix}_mean']
        # ROC-AUC поддерживает sample_weight
        auc_w = roc_auc_score(y_true_clf, y_score, sample_weight=weights)
        # бинаризация по порогу 0.5, затем precision/recall/f1 с весами
        y_pred_label = (y_score >= 0.5).astype(int)
        prec_w = precision_score(y_true_clf, y_pred_label, sample_weight=weights, zero_division=0)
        rec_w  = recall_score(y_true_clf, y_pred_label, sample_weight=weights, zero_division=0)
        f1_w   = f1_score(y_true_clf, y_pred_label, sample_weight=weights, zero_division=0)
        weighted_clf_metrics[w][prefix] = {
            'ROC_AUC': auc_w,
            'Precision': prec_w,
            'Recall': rec_w,
            'F1': f1_w
        }

# %% [markdown]
# **Шаг 8.** Выведем взвешенные метрики.

# %%
for w in weight_vars:
    print(f"\n=== Взвешенные метрики с весом `{w}` ===\n")
    print("Регрессия:")
    for prefix, mets in weighted_reg_metrics[w].items():
        print(f"  {prefix}: MSE={mets['MSE']:.4f}, RMSE={mets['RMSE']:.4f}, MAE={mets['MAE']:.4f}")
    print("\nКлассификация:")
    for prefix, mets in weighted_clf_metrics[w].items():
        print(f"  {prefix}: ROC_AUC={mets['ROC_AUC']:.4f}, "
              f"Precision={mets['Precision']:.4f}, Recall={mets['Recall']:.4f}, F1={mets['F1']:.4f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    roc_auc_score,
    precision_score,
    recall_score,
    f1_score, 
    accuracy_score
)
import numpy as np

# Assuming df is already loaded in the notebook

# Compute unweighted regression metrics
reg_metrics = {}
for prefix in prefixes:
    y_true = df['community_prediction']
    y_pred = df[f'{prefix}_mean']
    mse  = mean_squared_error(y_true, y_pred)
    rmse = mse ** 0.5
    mae  = mean_absolute_error(y_true, y_pred)
    reg_metrics[prefix] = {'MSE': mse, 'RMSE': rmse, 'MAE': mae}

# Compute weighted regression metrics
weight_vars = ['total_predictors']
weighted_reg_metrics = {w: {} for w in weight_vars}
for w in weight_vars:
    weights = df[w]
    for prefix in prefixes:
        y_true = df['community_prediction']
        y_pred = df[f'{prefix}_mean']
        mse_w = mean_squared_error(y_true, y_pred, sample_weight=weights)
        rmse_w = mse_w ** 0.5
        mae_w = mean_absolute_error(y_true, y_pred, sample_weight=weights)
        weighted_reg_metrics[w][prefix] = {'MSE': mse_w, 'RMSE': rmse_w, 'MAE': mae_w}

# Plot regression metrics comparisons
metrics = ['MSE', 'RMSE', 'MAE']
weight_labels = ['unweighted'] + weight_vars
all_reg = {'unweighted': reg_metrics, **weighted_reg_metrics}

prefixes_without_community = [p for p in prefixes if p != 'Community']

# Use custom colors for both regression and classification metrics
custom_colors = ['#c8a06e', '#af2c22']  # Yellow and Red

for metric in metrics:
    plt.figure(figsize=(8, 6))  # Increased figure size for better spacing
    n = len(prefixes_without_community)
    width = 0.2
    x = list(range(n))
    
    # Find best (minimum) value for each metric across all models
    best_values = []
    for i, w in enumerate(weight_labels):
        vals = [all_reg[w][p][metric] for p in prefixes_without_community]
        best_values.extend(vals)
    best_value = min(best_values)
    
    # Find which model achieves the best value
    best_model = None
    best_weight = None
    for w in weight_labels:
        for p in prefixes_without_community:
            if all_reg[w][p][metric] == best_value:
                best_model = p
                best_weight = w
    
    # Plot bars
    for i, w in enumerate(weight_labels):
        vals = [all_reg[w][p][metric] for p in prefixes_without_community]
        bars = plt.bar([xi + i*width for xi in x], vals, width=width, label=w, 
                color=custom_colors[i], edgecolor='black', linewidth=1, alpha=0.8)
        
        # Add value labels on top of bars
        for j, bar in enumerate(bars):
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                    f'{height:.3f}', ha='center', va='bottom', 
                    fontweight='bold', fontsize=8, rotation=45)
    
    # Add best performer line
    plt.axhline(y=best_value, color='black', linestyle='--', alpha=0.7)
    plt.text(n-0.5, best_value*1.02, f'{best_model} ({best_weight}): {best_value:.3f}', 
            color='black', fontweight='bold', ha='right')
    
    plt.xticks([xi + width/2*(len(weight_labels)-1) for xi in x], prefixes_without_community, fontsize=10, fontweight='bold')
    plt.ylabel(metric, fontsize=12, fontweight='bold')
    plt.title(f'Comparison of {metric} across Models and Weights', fontsize=14, fontweight='bold')
    plt.legend(title='Weight Type')
    plt.grid(axis='y', linestyle='--', alpha=0.3)
    
    # Add extra padding to the top to ensure labels don't get cut off
    plt.ylim(0, max(best_values) * 1.25)
    
    plt.tight_layout()
    plt.show()

# Compute classification metrics
clf_metrics = {}
weighted_clf_metrics = {w: {} for w in weight_vars}
for prefix in prefixes:
    y_true = df['answer'].astype(int)
    y_score = df[f'{prefix}_mean']
    # unweighted
    auc  = roc_auc_score(y_true, y_score)
    y_pred = (y_score >= 0.5).astype(int)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec  = recall_score(y_true, y_pred, zero_division=0)
    f1   = f1_score(y_true, y_pred, zero_division=0)
    accuracy = accuracy_score(y_true, y_pred)
    clf_metrics[prefix] = {'ROC_AUC': auc, 'Precision': prec, 'Recall': rec, 'F1': f1, 'Accuracy': accuracy}
    # weighted
    for w in weight_vars:
        weights = df[w]
        auc_w = roc_auc_score(y_true, y_score, sample_weight=weights)
        prec_w = precision_score(y_true, y_pred, sample_weight=weights, zero_division=0)
        rec_w  = recall_score(y_true, y_pred, sample_weight=weights, zero_division=0)
        f1_w   = f1_score(y_true, y_pred, sample_weight=weights, zero_division=0)
        accuracy_w = accuracy_score(y_true, y_pred, sample_weight=weights)
        weighted_clf_metrics[w][prefix] = {
            'ROC_AUC': auc_w, 'Precision': prec_w, 'Recall': rec_w, 'F1': f1_w, 'Accuracy': accuracy_w
        }

# Plot classification metrics comparisons
clf_metrics_all = {'unweighted': clf_metrics, **weighted_clf_metrics}
metrics_clf = ['ROC_AUC', 'Precision', 'Recall', 'F1', 'Accuracy']

x = list(range(len(prefixes)))

# Use the same custom colors for consistency
# custom_colors is already defined as ['#c8a06e', '#af2c22']

for metric in metrics_clf:
    plt.figure(figsize=(8, 6))
    
    # Find best (maximum) value for each metric across all models
    best_values = []
    for i, w in enumerate(weight_labels):
        vals = [clf_metrics_all[w][p][metric] for p in prefixes_without_community]
        best_values.extend(vals)
    best_value = max(best_values)
    
    # Find which model achieves the best value
    best_model = None
    best_weight = None
    for w in weight_labels:
        for p in prefixes_without_community:
            if clf_metrics_all[w][p][metric] == best_value:
                best_model = p
                best_weight = w
    
    # Plot bars
    for i, w in enumerate(weight_labels):
        vals = [clf_metrics_all[w][p][metric] for p in prefixes]
        bars = plt.bar([xi + i*width for xi in x], vals, width=width, label=w, 
                color=custom_colors[i], edgecolor='black', linewidth=1, alpha=0.8)
        
        # Add value labels on top of bars
        for j, bar in enumerate(bars):
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                    f'{height:.3f}', ha='center', va='bottom', 
                    fontweight='bold', fontsize=8, rotation=45)
    
    # Add best performer line
    plt.axhline(y=best_value, color='black', linestyle='--', alpha=0.7)
    plt.text(len(prefixes)-0.5, best_value*0.98, f'{best_model} ({best_weight}): {best_value:.3f}', 
            color='black', fontweight='bold', ha='right')
    
    plt.xticks([xi + width/2*(len(weight_labels)-1) for xi in x], prefixes, fontsize=10, fontweight='bold')
    plt.ylabel(metric, fontsize=12, fontweight='bold')
    plt.title(f'Comparison of {metric} across Models and Weights', fontsize=14, fontweight='bold')
    plt.legend(title='Weight Type')
    plt.grid(axis='y', linestyle='--', alpha=0.3)
    
    # Add extra padding to the top to ensure labels don't get cut off
    plt.ylim(0.2, 0.9)#min(1.1, best_value*1.2))
    
    plt.tight_layout()
    plt.show()
```

```python
# Текстовое представление метрик
def format_metrics(metrics_dict):
    lines = []
    for prefix, mets in metrics_dict.items():
        line = f"{prefix}: " + ", ".join(f"{k}={v:.4f}" for k, v in mets.items())
        lines.append(line)
    return "\n".join(lines)

# Выводим метрики регрессии
print("Регрессионные метрики:")
print(format_metrics(reg_metrics))

# Выводим взвешенные регрессионные метрики
for w in weight_vars:
    print(f"\n=== Взвешенные регрессионные метрики с весом `{w}` ===")
    print(format_metrics(weighted_reg_metrics[w]))
# Выводим классификационные метрики
print("\nКлассификационные метрики:")
print(format_metrics(clf_metrics))
# Выводим взвешенные классификационные метрики
for w in weight_vars:
    print(f"\n=== Взвешенные классификационные метрики с весом `{w}` ===")
    print(format_metrics(weighted_clf_metrics[w]))
    
```
