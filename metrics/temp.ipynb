```python
#drop HybridRAG_gd_1 column
df.drop(columns=["HybridRAG_gd_1", "HybridRAG_gd_1_text"], inplace=True)

#drop nans
df.dropna(inplace=True)
```

```python

df["Community_0"] = df["community_prediction"]

df.info()
```

```python
import pandas as pd
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    roc_auc_score,
    precision_score,
    recall_score,
    f1_score
)
```

```python
prefixes = {
    'PlainLLM': 3,
    'NaiveRAG_gd': 3,
    'NaiveRAG_lc': 3,
    'HybridRAG_gd': 1,
    'HybridRAG_lc': 3,
    'Community': 1,
}
```

```python
for prefix, count in prefixes.items():
    cols = [f'{prefix}_{i}' for i in range(count)]
    df[f'{prefix}_mean'] = df[cols].mean(axis=1)
    df[f'{prefix}_std']  = df[cols].std(axis=1)

df
```

```python
print(f"Mean std for NaiveRAG_gd: {df['NaiveRAG_gd_std'].mean()}")
print(f"Mean std for NaiveRAG_lc: {df['NaiveRAG_lc_std'].mean()}")
print(f"Mean std for PlainLLM: {df['PlainLLM_std'].mean()}")
print(f"Mean std for HybridRAG_lc: {df['HybridRAG_lc_std'].mean()}")
print()
print(f"Mean mean for NaiveRAG_gd: {df['NaiveRAG_gd_mean'].mean()}")
print(f"Mean mean for NaiveRAG_lc: {df['NaiveRAG_lc_mean'].mean()}")
print(f"Mean mean for PlainLLM: {df['PlainLLM_mean'].mean()}")
print(f"Mean mean for HybridRAG_lc: {df['HybridRAG_lc_mean'].mean()}")
print(f"Mean mean for HybridRAG_gd: {df['HybridRAG_gd_mean'].mean()}")


```

```python
reg_metrics = {}
for prefix in prefixes:
    y_true = df['community_prediction']
    y_pred = df[f'{prefix}_mean']
    mse  = mean_squared_error(y_true, y_pred)
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = mse ** 0.5
    reg_metrics[prefix] = {'MSE': mse, 'RMSE': rmse, 'MAE': mae}

```

```python
clf_metrics = {}
for prefix in prefixes:
    y_true = df['answer'].astype(int)
    y_score = df[f'{prefix}_mean']
    auc = roc_auc_score(y_true, y_score)
    y_pred_label = (y_score >= 0.5).astype(int)
    prec = precision_score(y_true, y_pred_label, zero_division=0)
    rec  = recall_score(y_true, y_pred_label, zero_division=0)
    f1   = f1_score(y_true, y_pred_label, zero_division=0)
    clf_metrics[prefix] = {
        'ROC_AUC': auc,
        'Precision': prec,
        'Recall': rec,
        'F1': f1
    }
```

```python
print("Регрессионные метрики против community_prediction:")
for prefix, mets in reg_metrics.items():
    print(f"  {prefix}: MSE={mets['MSE']:.4f}, RMSE={mets['RMSE']:.4f}, MAE={mets['MAE']:.4f}")

print("\nКлассификационные метрики против answer:")
for prefix, mets in clf_metrics.items():
    print(f"  {prefix}: ROC_AUC={mets['ROC_AUC']:.4f}, "
          f"Precision={mets['Precision']:.4f}, Recall={mets['Recall']:.4f}, F1={mets['F1']:.4f}")







```

```python
# %% [markdown]
# **Шаг 7.** Рассчитаем взвешенные метрики регрессии и классификации:
# - Регрессия: MSE, RMSE, MAE с учётом sample_weight  
# - Классификация: ROC-AUC, Precision, Recall, F1 с учётом sample_weight  
# для двух видов весов: `total_predictors` и `mean_ACE`.

# %%
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    roc_auc_score,
    precision_score,
    recall_score,
    f1_score
)

weight_vars = ['total_predictors', 'mean_ACE']
weighted_reg_metrics = {w: {} for w in weight_vars}
weighted_clf_metrics = {w: {} for w in weight_vars}

for w in weight_vars:
    weights = df[w]
    for prefix in prefixes:
        # --- Регрессия ---
        y_true = df['community_prediction']
        y_pred = df[f'{prefix}_mean']
        mse_w = mean_squared_error(y_true, y_pred, sample_weight=weights)
        rmse_w = mse_w ** 0.5
        mae_w = mean_absolute_error(y_true, y_pred, sample_weight=weights)
        weighted_reg_metrics[w][prefix] = {
            'MSE': mse_w,
            'RMSE': rmse_w,
            'MAE': mae_w
        }
        # --- Классификация ---
        y_true_clf = df['answer'].astype(int)
        y_score = df[f'{prefix}_mean']
        # ROC-AUC поддерживает sample_weight
        auc_w = roc_auc_score(y_true_clf, y_score, sample_weight=weights)
        # бинаризация по порогу 0.5, затем precision/recall/f1 с весами
        y_pred_label = (y_score >= 0.5).astype(int)
        prec_w = precision_score(y_true_clf, y_pred_label, sample_weight=weights, zero_division=0)
        rec_w  = recall_score(y_true_clf, y_pred_label, sample_weight=weights, zero_division=0)
        f1_w   = f1_score(y_true_clf, y_pred_label, sample_weight=weights, zero_division=0)
        weighted_clf_metrics[w][prefix] = {
            'ROC_AUC': auc_w,
            'Precision': prec_w,
            'Recall': rec_w,
            'F1': f1_w
        }

# %% [markdown]
# **Шаг 8.** Выведем взвешенные метрики.

# %%
for w in weight_vars:
    print(f"\n=== Взвешенные метрики с весом `{w}` ===\n")
    print("Регрессия:")
    for prefix, mets in weighted_reg_metrics[w].items():
        print(f"  {prefix}: MSE={mets['MSE']:.4f}, RMSE={mets['RMSE']:.4f}, MAE={mets['MAE']:.4f}")
    print("\nКлассификация:")
    for prefix, mets in weighted_clf_metrics[w].items():
        print(f"  {prefix}: ROC_AUC={mets['ROC_AUC']:.4f}, "
              f"Precision={mets['Precision']:.4f}, Recall={mets['Recall']:.4f}, F1={mets['F1']:.4f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    roc_auc_score,
    precision_score,
    recall_score,
    f1_score, 
    accuracy_score
)

# Assuming df is already loaded in the notebook

# Compute unweighted regression metrics
reg_metrics = {}
for prefix in prefixes:
    y_true = df['community_prediction']
    y_pred = df[f'{prefix}_mean']
    mse  = mean_squared_error(y_true, y_pred)
    rmse = mse ** 0.5
    mae  = mean_absolute_error(y_true, y_pred)
    reg_metrics[prefix] = {'MSE': mse, 'RMSE': rmse, 'MAE': mae}

# Compute weighted regression metrics
#weight_vars = ['total_predictors', 'mean_ACE']
weight_vars = ['total_predictors']
weighted_reg_metrics = {w: {} for w in weight_vars}
for w in weight_vars:
    weights = df[w]
    for prefix in prefixes:
        y_true = df['community_prediction']
        y_pred = df[f'{prefix}_mean']
        mse_w = mean_squared_error(y_true, y_pred, sample_weight=weights)
        rmse_w = mse_w ** 0.5
        mae_w = mean_absolute_error(y_true, y_pred, sample_weight=weights)
        weighted_reg_metrics[w][prefix] = {'MSE': mse_w, 'RMSE': rmse_w, 'MAE': mae_w}

# Plot regression metrics comparisons
metrics = ['MSE', 'RMSE', 'MAE']
weight_labels = ['unweighted'] + weight_vars
all_reg = {'unweighted': reg_metrics, **weighted_reg_metrics}

for metric in metrics:
    plt.figure(figsize=(10, 6))
    n = len(prefixes)
    width = 0.2
    x = list(range(n))
    for i, w in enumerate(weight_labels):
        vals = [all_reg[w][p][metric] for p in prefixes]
        plt.bar([xi + i*width for xi in x], vals, width=width, label=w)
    plt.xticks([xi + width for xi in x], prefixes)
    plt.ylabel(metric)
    plt.title(f'Comparison of {metric} across prefixes and weights')
    plt.legend()
    plt.show()

# Compute classification metrics
clf_metrics = {}
weighted_clf_metrics = {w: {} for w in weight_vars}
for prefix in prefixes:
    y_true = df['answer'].astype(int)
    y_score = df[f'{prefix}_mean']
    # unweighted
    auc  = roc_auc_score(y_true, y_score)
    y_pred = (y_score >= 0.5).astype(int)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec  = recall_score(y_true, y_pred, zero_division=0)
    f1   = f1_score(y_true, y_pred, zero_division=0)
    accuracy = accuracy_score(y_true, y_pred)
    clf_metrics[prefix] = {'ROC_AUC': auc, 'Precision': prec, 'Recall': rec, 'F1': f1, 'Accuracy': accuracy}
    # weighted
    for w in weight_vars:
        weights = df[w]
        auc_w = roc_auc_score(y_true, y_score, sample_weight=weights)
        prec_w = precision_score(y_true, y_pred, sample_weight=weights, zero_division=0)
        rec_w  = recall_score(y_true, y_pred, sample_weight=weights, zero_division=0)
        f1_w   = f1_score(y_true, y_pred, sample_weight=weights, zero_division=0)
        weighted_clf_metrics[w][prefix] = {
            'ROC_AUC': auc_w, 'Precision': prec_w, 'Recall': rec_w, 'F1': f1_w, 'Accuracy': accuracy
        }

# Plot classification metrics comparisons
clf_metrics_all = {'unweighted': clf_metrics, **weighted_clf_metrics}
metrics_clf = ['ROC_AUC', 'Precision', 'Recall', 'F1', 'Accuracy']

for metric in metrics_clf:
    plt.figure(figsize=(10, 6))
    for i, w in enumerate(weight_labels):
        vals = [clf_metrics_all[w][p][metric] for p in prefixes]
        plt.bar([xi + i*width for xi in x], vals, width=width, label=w)
    plt.xticks([xi + width for xi in x], prefixes)
    plt.ylabel(metric)
    plt.title(f'Comparison of {metric} across prefixes and weights')
    plt.legend()
    plt.show()
```
